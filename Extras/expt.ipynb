```python
# creating a sliding window of size 15 for each word in the sequence
def batch(l, b=15):
    subs = []
    for i in range(len(l) - b):
        subs.append(l[i : i + b])
    return subs

# create a incresing sized window for each word in the sequence starting from 10, 11, 12, 13, 14, 15, ...
def subdiv(l, b=10):
    subs = []
    for i in range(len(l) - b):
        subs.append(l[: i + b])  
    return subs
```

- if the number of tokens <= 23  `No repeatation`
- if the eos token is present in the sequence `No repeatation`


```python
for b in range(1):
    mask = indices[b] != 0
    N = mask.sum().item()
    var = np.array([np.var(s) / len(s) for s in batch(values[b, mask].float().numpy())])

    if len(var) < 10: # since window size is 15, hence if number of tokens <= 23, then no repetitions 
        output["repeats"].append(None)
        continue
    if (
        indices[b] == 6  # if any of the word is eos and (the number of mask tokens + eos token) is less than total number of tokens
    ).any() and N + 1 < indices.shape[1]: 
        # there is an end to the generation, likely no repetitions
        output["repeats"].append(None)
        continue
    
    varvar = np.array([np.var(v) for v in subdiv(var[::-1])][::-1])
    minlen = 120
    small_var = np.where(varvar < 0.045)[0] # returns the indices of the varvar array where the value is less than 0.045
    if len(small_var) > 1:
        if np.all(np.diff(small_var) < 2):
            idx = int(min(max(small_var[0], 1) * 1.08 + minlen, 4095))
            if idx / N > 0.9:  # at most last bit   # repeatation in the last 10% tokens of the sequence
                output["repeats"].append(None)
                continue
            elif small_var[0] < 30:
                idx = 0
            print("Found repetitions in sample %i" % b)
            output["repeats"].append(idx)
            output["sequences"][b, idx:] = 0
            output["repetitions"][b, :idx] = 0
        else:
            output["repeats"].append(None)
    else:
        output["repeats"].append(None)
```
