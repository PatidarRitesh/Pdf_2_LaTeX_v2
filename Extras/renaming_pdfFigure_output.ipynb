```python
import os
import subprocess
```

```python


def rename_files(directory_path, old_prefix, new_prefix):
    # List all files in the directory
    files = os.listdir(directory_path)

    for file_name in files:
        # Check if the file name starts with the old prefix
        if file_name.startswith(old_prefix):
            # Construct the new file name by replacing the old prefix with the new prefix
            new_name = file_name.replace(old_prefix, new_prefix)

            # Construct the full paths for the old and new names
            old_path = os.path.join(directory_path, file_name)
            # print("old_path :" ,old_path)
            new_path = os.path.join(directory_path, new_name)
            # print("new_path :" ,new_path)


            # Rename the file
            try:

                os.rename(old_path, new_path)
                print(f"Renamed: {file_name} to {new_name}")

            except Exception as e:
                print("Error: ", e)
                continue
            


```

```python

directory_path = '/home/patidarritesh/Nougat/pdffigures2/figure/data'
old_prefix = 'output'
new_prefix = ''
rename_files(directory_path, old_prefix, new_prefix)
```
