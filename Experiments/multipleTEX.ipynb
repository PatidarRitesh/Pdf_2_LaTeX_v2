## finding the name of root file or main.tex file 
##### * Because here if there are multiple .tex file then we can not say that there is always main.tex file named root file exist.
##### * We also find that there are exist more than 2 root .tex file . in this case we will ignore that files because we dont know which file is our main file to merge it.
#### for this task we used disjoint set (union) to find parent file.

```python
# given all the directories print the length if root_files for each directory

for directory in directories:
    ds=DisjointSet()
    count = 0
    for file in os.listdir(directory):
        if file.endswith('.tex'):
            count+=1
            with open(os.path.join(directory,file), 'rb') as f:
                try:
                    str=f.read()
                    str = str.decode("utf-8", errors="ignore")  # Ignore non-UTF-8 characters
                    res=re.findall(pattern,str)
                    for i in res:
                        if i[1] != '':
                            if any([1  for file in os.listdir(directory) if file.startswith(i[1])]):
                                ds.union(i[1],file)
                        elif i[3] != '':
                            if any([1  for file in os.listdir(directory) if file.startswith(i[3])]):
                                ds.union(i[3],file)
                except Exception as e:
                    print("Error in file: ",file, ": ", directory)
                    print(e)
                    print()

    ds=dict(ds)
    root_files=set(dict(ds).values())
    root_files=list(root_files)
    root_files.extend([file_1 for file_1 in os.listdir(directory) if file_1.endswith('.tex') and file_1.split(".tex")[0] not in ds.keys() and file_1 not in root_files])
    print(directory," : ",len(root_files), "total files", count )
    # print(ds)
    print(root_files)
    print("--------------------------------------------")
```
